```python
test1 = df.copy()
```

```python
test1.drop(['PropertyID', 'State', 'Block', 'Address'], axis=1, inplace=True)
```

## Encode Categorical Values and Split Data for the Model

```python
test1.head(5)
```

```python
cols_to_encode = ['Surroundings','BldgClassCategory','BldgClass_AtEvaluationTime']

for col in cols_to_encode:
    temp = pd.get_dummies(col,drop_first=True)
    test1.drop(col,axis=1,inplace=True)
    test1 = pd.concat([test1,temp],axis=1)
```

```python
test1
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
cols_to_label = ['ZipCode','Lot','YearOfConstruction']

for col in cols_to_label:
    test1[col] = le.fit_transform(test1[col])
```

```python
test1.head(5)
```

Create First Model

```python
X = test1.drop('PropertyEvaluationvalue',axis=1)
y = test1['PropertyEvaluationvalue']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test  = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
lin_rg = LinearRegression()
rdg_rg = Ridge()
lss_rg = Lasso()
dt_rg = DecisionTreeRegressor()
rf_rg = RandomForestRegressor()
```

```python
regressors = [lin_rg,rdg_rg,lss_rg,dt_rg,rf_rg]

for reg in regressors:
    reg.fit(x_train,y_train)
    pred = reg.predict(x_test)
    print("The Mean Squared Error score for", reg, "is", mean_squared_error(y_test,pred,squared=False))
    print("The R Squared Score for", reg, "is", r2_score(y_test,pred))
```
