It seems that we can drop a few features to start such as PropertyID, State, Block, Address to start with

```python
test1 = df.copy(deep=True)
```

```python
test1.drop(['PropertyID', 'State', 'Block', 'Address'], axis=1, inplace=True)
```

## Encode Categorical Values and Split Data for the Model

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for i in test1.columns:
    if test1[i].dtype == 'object':
        test1[i] = le.fit_transform(test1[i])
```

```python
X = test1.drop('PropertyEvaluationvalue', axis=1)
y = test1['PropertyEvaluationvalue']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

Some candidate features to drop are: Total Number of Units, Tax Class and Building Class at Evaluation, and Borough

```python
x_train = x_train.drop(['Borough','TotalNoOfUnits','TaxClass_AtEvaluationTime','BldgClass_AtEvaluationTime'], axis=1)
```

```python
x_test = x_test.drop(['Borough','TotalNoOfUnits','TaxClass_AtEvaluationTime','BldgClass_AtEvaluationTime'], axis=1)
```
